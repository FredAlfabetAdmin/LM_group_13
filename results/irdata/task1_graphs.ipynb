```python
import matplotlib.pyplot as plt
import numpy as np

file_path = "C:\\Users\\sando\\OneDrive\\Courses\\Master\\Learning_machines\\github\\LM_group_13\\results\\irdata\\task1_software_data.txt"

lines = open(file_path, 'r').readlines()

# Extract data and obstacle information
data = []
obstacle_indices = []
sensors = ['BackL', 'BackR', 'FrontL', 'FrontR', 'FrontC', 'FrontRR', 'BackC', 'FrontLL']
dif = 0

for i, line in enumerate(lines[0:]):  # Skip the header
    values = [float(val) if val.replace('.', '', 1).isdigit() or val.isdigit() else None for val in line.split()]
    if len(values) == 8:
        data.append(values)
    elif "Obstacle detected! Adjusting direction..." in line:
        obstacle_indices.append(i-1-dif)
    else:
        dif += 2

# Convert data to NumPy array
data = np.array(data)
# Extract time axis
time_axis = np.arange(len(data))

# Plotting
fig, ax = plt.subplots(figsize=(10, 8))

# Plot all sensor readings
for i, sensor_data in enumerate(data.T):
    ax.plot(time_axis, sensor_data, label=f"{sensors[i]}")

# Mark obstacles with blue points
ax.scatter(obstacle_indices,[2] *len(obstacle_indices), c='blue', marker='o', label='Obstacle Detected')

ax.set_xlabel('Time')
ax.set_ylabel('Sensor Reading')
ax.set_ylim(0, 200)
ax.set_xlim(10, 30)

ax.legend()
ax.grid(True)
plt.title('Simulation - IR data avoiding obstacle')
plt.savefig("results/figures/simulation_obstacle_irs.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (10, 5)
df = pd.read_csv('./trainings/higher_learning_rate/stats.csv')
mean_loss = df.groupby('epoch')['losses'].mean().reset_index()
median_loss = df.groupby('epoch')['losses'].median().reset_index()
min_loss = df.groupby('epoch')['losses'].min().reset_index()
max_loss = df.groupby('epoch')['losses'].max().reset_index()
std_loss = df.groupby('epoch')['losses'].std().reset_index()
plt.plot(median_loss['epoch'], median_loss['losses'], label='median')
plt.plot(mean_loss['epoch'], mean_loss['losses'], label='mean')
# plt.plot(min_loss['epoch'], min_loss['losses'], label='min')
# plt.plot(max_loss['epoch'], max_loss['losses'], label='max')
# plt.fill_between(mean_loss['epoch'], mean_loss['losses'] + std_loss['losses'], mean_loss['losses'] - std_loss['losses'], alpha=0.3)
plt.legend()
plt.tight_layout()
```
